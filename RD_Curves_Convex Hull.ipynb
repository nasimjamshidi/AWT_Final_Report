```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
clip = pd.read_csv("clip.csv") 
encode = pd.read_csv("encode.csv") 
```

##### Drop unwanted colomns from  the table Clip


```python
clip = clip.drop(["duration", "video_codec", "video_profile","format_long_name","sample_aspect_ratio","display_aspect_ratio","created_at","updated_at"], axis=1)
clip.head()z
```

##### Add prefix before each of the coloumn names in the table Clip

```python
clip=clip.add_prefix('clip_')
clip.head()
```

##### Get the last part of each file_name in the table Clip (Since the names of the files to some extent are the same,we only need the last part of each name to be seperatable) 
e.g.-> /Volumes/pte/content/upload/watchfolder/touchdown_pass_1080p_crf20.mp4 -> touchdown_pass_1080p_crf20.mp4

```python

clip= clip.dropna(subset=['clip_filename'])
size=clip['clip_filename'].shape[0]
for i in range(size):
    txt=clip['clip_filename'][i]
    out = txt.split("/")
    clip['clip_filename'][i]=out[-1]
```

##### Drop unwanted colomns from encode table

```python
encode = encode.drop(["created_at","updated_at","gop_size","b_frame_interval","frame_rate","reference_frames","b_frame_interval","stage_percentage","state"], axis=1)
encode.head()
```

##### Add prefix before each of the coloumn names in the table Encode

```python
encode=encode.add_prefix('encode_')
encode.head()
```

##### Get the last part of each file_name in the table encode

```python
encode= encode.dropna(subset=['encode_filename'])
size=encode['encode_filename'].shape[0]
for i in range(6340):
    txt=encode['encode_filename'][i]z
    out = txt.split("/")
    encode['encode_filename'][i]=out[-1]
```

##### Merge two tables Clip and Encode


```python
encode_clip=encode.merge(clip, left_on='encode_clip_id', right_on='clip_id',how='left')
encode_clip.head()
```

#####  Get the unique values in the clip_filename col

```python
unq=encode_clip['clip_filename'].unique()
```

##### Convert Bitrate values from KBPS to MBPS

```python
for i in range(len(encode_clip['encode_bitrate_video'])):
    encode_clip['encode_bitrate_video'][i] = 1/1000 * encode_clip['encode_bitrate_video'][i]
```

##### Fit the Convex Hull on RD curves of 20 different videos based on PSNR metric
